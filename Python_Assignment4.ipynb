```python
#LINE GRAPH
X = [5,10,15,20,25,30,35,40]
Y = [76,33,22,67,99,15,43,67]
Z = [76,45,39,87,24,67,46,89]

plt.xlim(0,50)
plt.ylim(0,120)

plt.text(40,20,"C0894123",fontsize=15,color='#1765F8')

plt.plot(X,Y,label='Punjab',linewidth=3,color='red',marker='o',markersize=10,markerfacecolor='w')
plt.plot(X,Z,label='Haryana',linewidth=3,color='k',marker='o',markersize=10,markerfacecolor='w',linestyle='dotted')

plt.title("Suicides in North India",fontsize=15,color='#1765F8')
plt.xlabel("Year",fontsize=15,color='#000000')
plt.ylabel("Total Sucides",fontsize=15,color='black')

plt.legend()
plt.show()
```

```python
# SCATTER PLOT
X = [5,10,15,20,25,30,35,40]
Y = [76,33,22,67,99,15,43,67]
Z = [76,45,39,87,24,67,46,89]

plt.xlim(0,50)
plt.ylim(0,120)
plt.scatter(X, Y, color='brown', label='Punjab')
plt.scatter(X, Z, color='yellow', label='Haryana')

plt.title('Total Suicides in India')
plt.xlabel('Year 20**')
plt.ylabel('Total Suicides')

plt.legend()
plt.show()
```

```python
#PIE CHART
plt.title("Day in my Life")
hrs = [8,8,2,2,1,3]
cols = ['brown','green','b','cyan','y','purple']
activity = ['Sleeping','Work','Commute','Eating','Miscellaneous','Studying']
plt.pie(hrs,colors=cols,labels=activity,autopct='%.2f%%')

plt.legend(fontsize=12,bbox_to_anchor=(1.65,0.8))
plt.show()
```

```python
# EXPLODED PIE CHART
plt.title("Day in my Life: Exploded Pie Chart")
hrs = [8,8,2,2,1,3]
cols = ['brown','green','b','cyan','y','purple']
activity = ['Sleeping','Work','Commute','Eating','Miscellaneous','Studying']
explode = (0.03, 0.05, 0.02, 0.09,0.05,0.01)
plt.pie(hrs,colors=cols,labels=activity,explode=explode,autopct='%.2f%%')

plt.legend(fontsize=12,bbox_to_anchor=(1.65,0.8))
plt.show()

```

```python
#HISTOGRAM
data = np.random.randn(1000)

plt.hist(data, bins=30, color='cyan', edgecolor='blue', alpha=0.7, label='Histogram')

plt.title('Histogram')
plt.xlabel('City ID')
plt.ylabel('No of Covid Patients')

plt.legend()
plt.show()
```

```python
#BOX PLOT
data = np.random.randn(100)
ax = plt.axes()
ax.set_facecolor("lightblue")
plt.boxplot(data, vert=False, patch_artist=True, labels=['Box Plot'])
plt.title('Box Plot')
plt.show()
```

```python
# Error bar plot
x = np.array([1, 2, 3, 4, 5])
y = x**2
y_error = np.array([1, 0.5, 1.5, 1, 2])
ax = plt.axes()
ax.set_facecolor("lightblue")

plt.errorbar(x, y, yerr=y_error, fmt='o', color='blue', label='Error Bar Plot')
plt.title('Error Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.show()
```
